```python
import os
from PIL import Image
import re
import matplotlib.image as mpimg
```

```python
source_path = '/content/drive/MyDrive/capstone/the-data/data_raw'
clean_path = '/content/drive/MyDrive/capstone/the-data/clean'
folder = os.listdir(source_path)
folder
```

```python
for category in folder:
  os.mkdir(os.path.join(clean_path, category))
```

```python
for category in os.listdir(source_path):
    source_path_per_category = os.path.join(source_path, category)
    clean_path_per_category = os.path.join(clean_path, category)

    # Ensure the clean path per category exists
    if not os.path.exists(clean_path_per_category):
        os.makedirs(clean_path_per_category)

    list_files_per_category = os.listdir(source_path_per_category)

    # Process each image in the category folder
    for image in list_files_per_category:
        image_file = os.path.join(source_path_per_category, image)
        im = Image.open(image_file)

        # Regex to select only number in filename
        number = re.search(r"^\d+", image)

        # Handle cases where no number is found
        if number:
            clean_file = number.group(0) + ".jpg"
        else:
            # Use the original name without extension if no number is found
            clean_file = os.path.splitext(image)[0] + ".jpg"

        # Resize image and convert to .jpg
        file_save = os.path.join(clean_path_per_category, clean_file)
        im_final = im.resize((300, 300))
        im_final = im_final.convert("RGB")

        print(f"Saving file to: {file_save}")
        im_final.save(file_save)
```
